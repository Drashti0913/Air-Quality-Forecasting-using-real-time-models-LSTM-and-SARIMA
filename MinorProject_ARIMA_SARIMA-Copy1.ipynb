```python
from statsmodels.tsa.stattools import adfuller
```

```python
test_result= adfuller(df['PM2.5'])
```

```python
#H0 = It is not Stationary
#H1 = It is Stationary

def adfuller_test(PM25):
    result = adfuller(PM25)
    labels = ['ADF Test Statistic','p-value','Lags Used','Number of Observation Used']
    for value,label in zip(result,labels):
        print(label+ " : " +str(value))
    if result[1] <= 0.05:
        print("Strong Evidence against the null hypothesis(H0), reject the null hypothesis. Data has no unit root and it is stationary")
    else:
        print("Weak evidence against null hypothesis, time series has a unit root, indicating it is not stationary")
    
```

```python
adfuller_test(df['PM2.5'])
```

```python
from statsmodels.tsa.arima.model import ARIMA
#p,d,q --> p= AR model lags, q = MA lags 0 or 1, d = differencing
# Fit ARIMA model
model_arima = ARIMA(df['PM2.5'], order=(1,1,1))
model_arima_fit = model_arima.fit()

# Display model summary
print(model_arima_fit.summary())

# Predict using ARIMA
df['arima_forecast'] = model_arima_fit.predict(start=90, end=103, dynamic=True)
df[['PM2.5', 'arima_forecast']].plot(figsize=(12,8))
plt.title('ARIMA Forecast vs Actuals')
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit SARIMA model
model_sarima = SARIMAX(df['PM2.5'], order=(1,1,1), seasonal_order=(1,1,1,12))
model_sarima_fit = model_sarima.fit()
#p,q,d,season (how many u are shifting)
# Display model summary
print(model_sarima_fit.summary())

# Predict using SARIMA
df['sarima_forecast'] = model_sarima_fit.predict(start=90, end=103, dynamic=True)
df[['PM2.5', 'sarima_forecast']].plot(figsize=(12,8))
plt.title('SARIMA Forecast vs Actuals')
plt.show()

```

```python
from pandas.tseries.offsets import DateOffset

# Generate future dates
future_dates = [df.index[-1] + DateOffset(months=i) for i in range(0,24)]
further_dates_df = pd.DataFrame(index=future_dates[1:], columns=df.columns)

# Concatenate the original dataframe with the future dates dataframe
future_df = pd.concat([df, further_dates_df])

# Forecast for future dates using SARIMA
forecast = model_sarima_fit.get_forecast(steps=24)
future_df['future_forecast'] = pd.concat([df['PM2.5'], forecast.predicted_mean])

# Plot the actual values and the forecast
future_df[['PM2.5', 'future_forecast']].plot(figsize=(12,10))
plt.title('SARIMA Future Forecast')
plt.show()
```
